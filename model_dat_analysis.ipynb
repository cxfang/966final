```python
velocities = ['slow', 'med', 'fast']
frictions = ['high', 'med', 'low']
fr_noise = 0.1
vel_noise = 0.05
col_noise = 0.15
df = pd.DataFrame()
for v in velocities:
    for f in frictions:
        data = np.load('model_data/model3/{}_{}_fr{}_vel{}_col{}.npy'.format(v, f, fr_noise, vel_noise, col_noise), allow_pickle=True).item()
        new_df = pd.DataFrame()
        colls = data.keys()
        probs = []
        vals = data.values()
        for val in vals:
            probs.append(np.mean(val))

        new_df['collider_position'] = colls
        new_df['model_prediction'] = probs

        new_df['friction'] = f
        new_df['speed'] = v
        new_df['fr_noise'] = fr_noise
        new_df['vel_noise'] = vel_noise
        new_df['col_noise'] = col_noise

        df = pd.concat([df, new_df])
df
```

```python
df.to_csv('analysis/model_data/model3/m3_fr{}_vel{}_col{}.csv'.format(fr_noise, vel_noise, col_noise), index=False)
# df.to_csv('model_data/preds/{}_{}_noise{}.csv'.format(speed, friction, fr_noise), index=False)
```
